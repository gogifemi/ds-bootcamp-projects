## Exploring and cleaning the data

To test our hypotheses and address some of the agency's key questions, we will now explore the data in detail.

**Dropping columns**  
As our agency has specific queries regarding the data, not all columns are relevant to our analysis. Consequently, we can exclude certain columns to optimise the exploration process.

```python
# drop columns we don't need
df.drop(["tags", "thumbnail_link", "comments_disabled", "ratings_disabled", "video_error_or_removed", "description"], axis=1, inplace=True)
```

```python
# check result
df.head()
```

**Dealing with duplicate rows**  
As stated in [this analysis](https://ammar-alyousfi.com/2020/youtube-trending-videos-analysis-2019-us), we will conduct our analysis on all 40,840 trending video records, rather than focusing only on unique entries. This approach ensures that videos trending for multiple days receive greater weight, aligning with our objectives.

```python
# check for duplicate rows in video_id column
df["video_id"].duplicated().value_counts()
```

Our third hypothesis has been validated: the dataset contains meaningful duplicates.

**Transforming data types**  
As mentioned earlier, we need to convert the `trending_date` and `publish_time` columns into a datetime format.    
We can use pandas' `to_datetime()` function to perform this transformation.

```python
# convert trending_date and publish_time into datetime and extract date part from publish_time
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
df['publish_time'] = pd.to_datetime(df['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
df['publish_date'] = pd.to_datetime(df['publish_time'].dt.date)
```

```python
# check result
df.head()
```

**Building new columns/ Feature engineering**  
In order to gain insight into the publication schedule of trending videos, it is essential to create new columns that provide data on the day of the week and time of day.

```python
# extract year, month, dayofweek and hour information out of column publish_time and build new column for each
df["publish_year"]=df["publish_time"].dt.year
df["publish_month"]=df["publish_time"].dt.month
df["publish_weekday"]=df["publish_time"].dt.dayofweek
df["publish_hour"]=df["publish_time"].dt.hour
```

```python
# check result
df.head()
```

```python
# extract year, month, day-of-week out of column trending_date and build new column for each
df["trending_year"]=df["trending_date"].dt.year
df["trending_month"]=df["trending_date"].dt.month
df["trending_weekday"]=df["trending_date"].dt.dayofweek
```

```python
# check result
df.head()
```

## Addressing the Business Case: Hypotheses Driven Approach

So far, we dealt with data so that we can easily analyse it. However, the dataset provided us with a valid and complete data, therefore the first and second hypthoses are correct. ## Hypotheses

1. Video-Categories differ in amount of trending videos, views, likes and dislikes.
2. For certain categories it takes longer to become a trending video after publication than for other categories.
3. Some YouTube channels are more successful than others.
4. The amount of video publications differ per day and hour.
5. Correlations exist between views, likes, dislikes and comments.
6. The length of time-interval between the date a video gets published and a video becomes trending also is correlated to number of views, likes, dislikes and comments.
7. Trending videos with higher engagement metrics (views, likes, comments) stay on the trending list for more days.

**7. Trending videos with higher engagement metrics (views, likes, comments) stay on the trending list for more days.**    

In order to test this hypothesis, it is essential to identify and count duplicate videos in order to examine the relationship between engagement metrics and the number of days spent on the trending list.

```python
# create a new column for displaying the duration on the trending list
trending_duration=df.groupby('video_id')['trending_date'].nunique()
df=df.merge(trending_duration, on='video_id')
```

```python
df=df.rename(columns={'trending_date_x':'trending_date', 'trending_date_y':'trending_duration'})
```

```python
# save the average of evaluation metrics
engagement_metrics=df.groupby('video_id')[['views', 'likes', 'dislikes', 'comment_count']].mean()
df=df.merge(engagement_metrics, on='video_id', suffixes=('', '_avg'))
```

```python
from scipy.stats import pearsonr
```

```python
# calculate the correlation between the engagement metrics and trending_duration
correlation_view, p_value_view = pearsonr(df['views_avg'], df['trending_duration'])
print(f"Correlation between views and trending days: {correlation_view}, p-value: {p_value_view}")

correlation_likes, p_value_likes = pearsonr(df['likes_avg'], df['trending_duration'])
print(f"Correlation between likes and trending days: {correlation_likes}, p-value: {p_value_likes}")

correlation_dislikes, p_value_dislikes = pearsonr(df['dislikes_avg'], df['trending_duration'])
print(f"Correlation between dislikes and trending days: {correlation_dislikes}, p-value: {p_value_dislikes}")

correlation_comments, p_value_comments = pearsonr(df['comment_count_avg'], df['trending_duration'])
print(f"Correlation between comments and trending days: {correlation_comments}, p-value: {p_value_comments}")
```

According to these results, it can be said that there are no positive or meaningful relationship between engagement metrics and video's duration on the trending list.
